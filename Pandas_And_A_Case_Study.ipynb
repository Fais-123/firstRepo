# __Case Study Titanic Dataset Using Python Libraries__ 

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
titanic=sns.load_dataset("titanic")
titanic.head(5) #viewing initial 5 rows of titanic dataset.
```

```python
#converting dataset into csv file
titanic.to_csv("titanic.csv")
```

```python
#to view summary of datasets
titanic.describe()
```

```python
titanic.head(5)
```

```python
#drop few columns from datset and making a new dataset
new=titanic.drop(["deck","embark_town","alone"],axis=1)  #axis=1 bcz we are dropping columns.
new
```

```python
titanic.mean() #mean of every column contains numeric value.
```

```python
#to get the count how many people srvived in titanic
titanic.value_counts(["survived"])
#result is 0 means false 549 dies and 1 means 342 remains alive.
```

```python
#using groupby
titanic.groupby(["sex"]).mean() #mean/average of quantites according to gender

```

```python
titanic.groupby(["sex","class"]).mean() #mean/average of quantites according to gender and class

```

```python
titanic.value_counts(["sex"]) #to get total count of males and females
```

```python
#children survived
titanic[titanic["age"] < 18].groupby(["sex","class"]).mean()

```
