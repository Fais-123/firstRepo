# **EXPLORATORY DATA ANALYSIS**

Creating a pie chart to see the distribution of top 3 zomato consumers over countires

```python
country_index = final_df['Country'].value_counts().index
```

```python
country_val = final_df['Country'].value_counts().values
```

```python
plt.pie(x=country_val[:3],labels=country_index[:3],autopct="%1.2f%%")
```

# **Observation:**
- India has the highest number of zomato consumers as compared to other countries.

Use Aggregate rating to find its rating text or status(e.g; good,poor)

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Frequency'})
ratings
```

# **Observation:**

- when rating is between 4.5 to 4.9 it is Excellent
- when rating is between 4.0 to 4.4 it is very Good
- when rating is between 3.5 to 3.9 it is  Good
- when rating is between 2.5 to 3.4 it is Average
- when rating is between 1.8 to 2.4 it is Poor
- 2148 records have 0 ratings. Not rated by the customers



# **Barplot for Visualization:**

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating Frequency',data=ratings)
```

- Top 5 cuisines with high Prices

```python
avg_price = final_df.groupby('Cuisines')['Price range'].mean()
avg_price.sort_values(ascending=False).head(5)
```
